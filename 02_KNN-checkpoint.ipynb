```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sklearn imports for modeling
from sklearn.model_selection import train_test_split, TimeSeriesSplit, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, confusion_matrix
from sklearn.pipeline import Pipeline
```

```python
RANDOM_STATE = 42
```

```python
try:
    df = pd.read_parquet("train-00000-of-00001.parquet")

    # Sort by Ticker and Date to ensure time-series integrity (past -> future)
    df = df.sort_values(['Ticker', 'Date']).reset_index(drop=True)

    features = [
        'Price_Change_5d',
        'RSI',
        'MACD_Histogram',
        'Volatility'
    ]

    target = 'Future_Up_5d'

    # Create the modeling dataset by dropping rows with NaNs in our specific columns
    df_model = df[features + [target]].dropna()

    X = df_model[features]
    y = df_model[target]

    print(f"Data Loaded. Shape: {df_model.shape}")
    print(f"Class Balance:\n{y.value_counts(normalize=True)}")


except FileNotFoundError:
    print("Error: The dataset file 'df_model_top3_by_sector.parquet' was not found.")
    # Add dummy data generation here if needed for testing code logic without file
```

```python
df_model.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, shuffle=False
)

print(f"Training samples: {len(X_train)} | Test samples: {len(X_test)}")
```

```python
log_reg_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('log_reg', LogisticRegression(
        random_state=RANDOM_STATE,
        solver='liblinear',      # Good for smaller datasets / binary classification
        class_weight='balanced',  # Penalizes mistakes on the minority class more
        # Regularization strength (smaller = stronger regularization to prevent overfitting)
        C=0.1
    ))
])

# Fit the model
print("Training Logistic Regression...")
log_reg_pipeline.fit(X_train, y_train)

# Predict
y_pred_lr = log_reg_pipeline.predict(X_test)
y_prob_lr = log_reg_pipeline.predict_proba(X_test)[:, 1]

# Evaluate
print("\n--- Logistic Regression Results ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_lr):.4f}")
print(f"ROC-AUC:  {roc_auc_score(y_test, y_prob_lr):.4f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred_lr))
```

```python
knn_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier())
])

# Define the grid of neighbors to test (odd numbers to avoid ties)
k_values = range(1, 30, 2)
param_grid = {'knn__n_neighbors': k_values}

# Use TimeSeriesSplit for Cross-Validation to respect temporal order
tscv = TimeSeriesSplit(n_splits=3)

print("Tuning KNN (this may take a moment)...")
grid_search = GridSearchCV(
    knn_pipeline,
    param_grid,
    cv=tscv,
    scoring='roc_auc',  # We optimize for AUC since accuracy can be misleading in finance
    n_jobs=-1           # Use all processors
)

grid_search.fit(X_train, y_train)

# Get best model
best_k = grid_search.best_params_['knn__n_neighbors']
best_knn_model = grid_search.best_estimator_

print(f"Best 'k' found: {best_k}")
print(f"Best Validation ROC-AUC: {grid_search.best_score_:.4f}")
```

```python
y_pred_knn = best_knn_model.predict(X_test)
y_prob_knn = best_knn_model.predict_proba(X_test)[:, 1]

print("\n--- KNN Results ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_knn):.4f}")
print(f"ROC-AUC:  {roc_auc_score(y_test, y_prob_knn):.4f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred_knn))
```

```python
models = ['Improved LogReg', f'KNN (k={best_k})']
scores = [roc_auc_score(y_test, y_prob_lr), roc_auc_score(y_test, y_prob_knn)]

plt.figure(figsize=(8, 5))
plt.bar(models, scores, color=['skyblue', 'salmon'])
plt.axhline(0.5, color='red', linestyle='--', label='Random Guess')
plt.ylabel('ROC-AUC Score')
plt.title('Model Comparison: Logistic Regression vs KNN')
plt.legend()
plt.ylim(0.4, 0.6)  # Zoom in on the relevant range for financial data
plt.show()
```
