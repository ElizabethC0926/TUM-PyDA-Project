## Section 2: Random Forest Model(Ensemble)
### 2.1: Motivation and model choice: Random Forest
Random Forest is chosen because it is a robust **ensemble** learning method that reduces variance through bagging and can capture non-linear relationships without strong parametric assumptions. These properties make it well suited for noisy financial data, where relationships between technical indicators and price movements are often unstable.

In addition, Random Forest requires minimal preprocessing, does not rely on feature scaling, and provides interpretable feature importance measures. For these reasons, it serves as a strong and widely used baseline model in both academic and industrial applications and provides a solid reference point for subsequent model improvements and comparisons.

### 2.2: Baseline Random Forest Model Setup
#### 2.2.1 First, define targest and baseline feature set.

**Target**: Future_Up_5d, 5-day direction (binary classification).

**Feature**: Price_Change_5d, 5-day price change (momentum, less noisy than 1-day change).

```python
# Target
Y = df["Future_Up_5d"].copy()

# Feature
X = df[["Price_Change_5d"]].copy()

# Drop missing values
df_model = pd.concat([X, Y], axis=1).dropna()

X = df_model[["Price_Change_5d"]]
Y = df_model["Future_Up_5d"]

print("Modeling data shape:", df_model.shape)
print("Class balance (Y):")
print(Y.value_counts(normalize=True))
```

#### 2.2.2: Train- test split(time aware)
With financial data, it is important to consider time when doing train/test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y,
    test_size=0.30,
    shuffle=False
)

print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
print("Train class balance:")
print(Y_train.value_counts(normalize=True))
print("Test class balance:")
print(Y_test.value_counts(normalize=True))
```

#### 2.2.3: Train baseline random forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=300,
    max_depth=5,
    min_samples_leaf=50,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, Y_train)
rf
```

#### 2.2.4: Evaluation

```python
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix

Y_pred = rf.predict(X_test)
Y_prob = rf.predict_proba(X_test)[:, 1]

accuracy = accuracy_score(Y_test, Y_pred)
roc_auc = roc_auc_score(Y_test, Y_prob)

print("Accuracy:", accuracy)
print("ROC-AUC:", roc_auc)
```

#### 2.2.5 Confusion matrix

```python
import matplotlib.pyplot as plt

cm = confusion_matrix(Y_test, Y_pred)

plt.figure()
plt.imshow(cm)
plt.title("Random Forest Confusion Matrix (Baseline)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.colorbar()
plt.show()

print("Confusion Matrix:\n", cm)
```

#### 2.2.6 Feature importance for single feature

```python
feature_importance = pd.Series(
    rf.feature_importances_,
    index=X.columns
).sort_values(ascending=False)

feature_importance
```

#### 2.2.7 Conclusion
The baseline Random Forest model achieves an accuracy of 55.2% and a ROC-AUC of 0.505 on the test set. While performance is marginally above a naive majority-class baseline, the improvement is very limited. 

The confusion matrix indicates that the model predominantly predicts upward movements, reflecting the slight class imbalance in the data. 

Feature importance assigns full weight to the 5-day price change, which is expected given the single-feature setup. Overall, the results suggest that medium-term momentum alone provides only weak predictive power for next-day stock price direction.

This leaves us room for improvement.

### 2.3: Adding CV
#### 2.3.1: Motivation
Standard K-Fold cross-validation is inappropriate for financial time series because it breaks temporal ordering and can introduce look-ahead bias. To obtain a more realistic estimate of out-of-sample performance, we apply time-series cross-validation using a **walk-forward (expanding window) scheme**.

#### 2.3.2: Method:
We use ***TimeSeriesSplit***, which preserves chronological order by training on past observations and validating on subsequent periods. Model performance is evaluated using Accuracy, ROC-AUC, and F1-score to assess both classification correctness and ranking quality.

```python
from sklearn.model_selection import TimeSeriesSplit, cross_validate
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Time-series cross-validation
tscv = TimeSeriesSplit(n_splits=5)

rf_cv = RandomForestClassifier(
    n_estimators=300,
    max_depth=5,
    min_samples_leaf=50,
    random_state=42,
    n_jobs=-1
)

scoring = {
    "accuracy": "accuracy",
    "roc_auc": "roc_auc",
    "f1": "f1"
}

cv_results = cross_validate(
    rf_cv,
    X, Y,
    cv=tscv,
    scoring=scoring,
    return_train_score=True
)

print("CV Accuracy:  {:.4f} ± {:.4f}".format(
    cv_results["test_accuracy"].mean(),
    cv_results["test_accuracy"].std()
))
print("CV ROC-AUC:  {:.4f} ± {:.4f}".format(
    cv_results["test_roc_auc"].mean(),
    cv_results["test_roc_auc"].std()
))
print("CV F1-score: {:.4f} ± {:.4f}".format(
    cv_results["test_f1"].mean(),
    cv_results["test_f1"].std()
))
```

#### 2.3.3 Conclusion
Time-series CV confirms the robustness of the Random Forest model across different temporal folds. The model achieves an average accuracy of 54.5% and a ROC-AUC of 0.503, with low variance across folds. These results indicate stable but limited predictive power, consistent with the inherent difficulty of short-term stock direction prediction. The relatively high F1-score reflects the slight class imbalance rather than strong discriminative ability, underscoring the importance of using multiple evaluation metrics.

### 2.4: Hyper parameter tuning
#### 2.4.1: Motivation
While Random Forest is robust to overfitting, its performance can still depend on key hyperparameters such as tree depth and minimum leaf size. To improve generalization and assess whether performance gains are achievable, we apply hyperparameter tuning using a time-series–aware validation strategy.

#### 2.4.2: Tuning strategy
A Randomized Search is applied to explore a limited, well-motivated hyperparameter space. The search objective is ROC-AUC, which evaluates ranking quality independent of a fixed decision threshold.

```python
from sklearn.model_selection import TimeSeriesSplit, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier

# Time-series cross-validation
tscv = TimeSeriesSplit(n_splits=5)

# Hyperparameter search space (conservative)
param_dist = {
    "n_estimators": [200, 300, 500],
    "max_depth": [3, 5, 7, None],
    "min_samples_leaf": [10, 20, 50, 100],
    "max_features": ["sqrt", 1.0]
}

rf_base = RandomForestClassifier(
    random_state=42,
    n_jobs=-1
)

search = RandomizedSearchCV(
    estimator=rf_base,
    param_distributions=param_dist,
    n_iter=15,
    scoring="roc_auc",
    cv=tscv,
    random_state=42,
    n_jobs=-1
)

search.fit(X, Y)

print("Best parameters:")
print(search.best_params_)

print("Best CV ROC-AUC:")
print(search.best_score_)
```

#### 2.4.3 Evaluation
The best-performing model from the tuning procedure is evaluated on the previously defined hold-out test set.

```python
best_rf = search.best_estimator_

Y_pred_tuned = best_rf.predict(X_test)
Y_prob_tuned = best_rf.predict_proba(X_test)[:, 1]

from sklearn.metrics import accuracy_score, roc_auc_score

accuracy_tuned = accuracy_score(Y_test, Y_pred_tuned)
roc_auc_tuned = roc_auc_score(Y_test, Y_prob_tuned)

print("Tuned Accuracy:", accuracy_tuned)
print("Tuned ROC-AUC:", roc_auc_tuned)
```

#### 2.4.4 Comparison with baseline model
Hyperparameter tuning leads to a measurable improvement in predictive performance. The tuned Random Forest achieves an accuracy of 55.2% and a ROC-AUC of 0.527 on the hold-out test set, outperforming the baseline model. The selected configuration favors shallow trees and strong regularization, indicating that limiting model complexity is essential when modeling noisy financial time series. While the overall predictive power remains modest, the results demonstrate that careful tuning can extract additional signal beyond the baseline configuration.

```python
from sklearn.model_selection import TimeSeriesSplit, cross_val_score
import numpy as np

tscv = TimeSeriesSplit(n_splits=5)

# Baseline RF
rf_baseline = RandomForestClassifier(
    n_estimators=300,
    max_depth=5,
    min_samples_leaf=50,
    random_state=42,
    n_jobs=-1
)

baseline_scores = cross_val_score(
    rf_baseline, X, Y,
    cv=tscv,
    scoring="roc_auc"
)

# Tuned RF (best estimator)
tuned_scores = cross_val_score(
    best_rf, X, Y,
    cv=tscv,
    scoring="roc_auc"
)

print("Baseline CV ROC-AUC:", baseline_scores)
print("Tuned CV ROC-AUC:", tuned_scores)

print("Baseline mean ± std:",
      baseline_scores.mean(), "±", baseline_scores.std())
print("Tuned mean ± std:",
      tuned_scores.mean(), "±", tuned_scores.std())
```

#### 2.4.5 Conclusion
A comparison of cross-validated ROC-AUC scores indicates that hyperparameter tuning results in a small but consistent improvement over the baseline model. The tuned model achieves a slightly higher average ROC-AUC (0.504) compared to the baseline (0.503) and exhibits lower variability across folds, suggesting improved stability and generalization.

Although the absolute performance gain is modest, the consistent improvement across validation folds indicates that stronger regularization helps mitigate noise in short-horizon stock prediction. Overall, the results suggest that hyperparameter tuning enhances robustness but that the model’s predictive power remains fundamentally constrained by the weak signal present in financial time series data.

## Section 4: Feature augmented RF
### 4.1 Motivation
We previously established a baseline Random Forest model and compared it against Logistic Regression using a fixed feature set. While Random Forest performs better than the linear baseline, overall predictability remains limited. We therefore investigate whether adding a small number of economically motivated technical indicators improves predictive performance. To preserve interpretability and avoid feature dumping, features are added incrementally.

### 4.3: Helper functions(train/eval + CV + Plots)

```python
def fit_eval_rf(X, Y, model_name, rf_params=None):
    """
    Train/test evaluation with chronological split + confusion matrix.
    Returns a dict of metrics and the trained model.
    """
    if rf_params is None:
        rf_params = {}

    # drop missing values (rolling features produce NaNs)
    tmp = pd.concat([X, Y], axis=1).dropna()
    X_clean = tmp[X.columns]
    Y_clean = tmp[Y.name]

    X_train, X_test, Y_train, Y_test = train_test_split(
        X_clean, Y_clean, test_size=0.30, shuffle=False
    )

    rf = RandomForestClassifier(
        random_state=RANDOM_STATE,
        n_jobs=-1,
        **rf_params
    )

    rf.fit(X_train, Y_train)

    Y_pred = rf.predict(X_test)
    Y_prob = rf.predict_proba(X_test)[:, 1]

    acc = accuracy_score(Y_test, Y_pred)
    auc = roc_auc_score(Y_test, Y_prob)
    cm = confusion_matrix(Y_test, Y_pred)

    print(f"\n=== {model_name}: Hold-out Test ===")
    print("Features:", list(X.columns))
    print("Accuracy:", acc)
    print("ROC-AUC:", auc)
    print("Confusion Matrix:\n", cm)

    # quick confusion matrix plot
    plt.figure()
    plt.imshow(cm)
    plt.title(f"Confusion Matrix - {model_name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.colorbar()
    plt.show()

    results = {
        "model": model_name,
        "n_features": X.shape[1],
        "features": list(X.columns),
        "test_accuracy": acc,
        "test_roc_auc": auc
    }

    return rf, results, X_clean, Y_clean


def timeseries_cv_rf(X, Y, model_name, rf_params=None, n_splits=5):
    """
    TimeSeriesSplit CV with mean ± std for Accuracy, ROC-AUC, F1.
    Returns dict of CV summary metrics.
    """
    if rf_params is None:
        rf_params = {}

    tmp = pd.concat([X, Y], axis=1).dropna()
    X_clean = tmp[X.columns]
    Y_clean = tmp[Y.name]

    tscv = TimeSeriesSplit(n_splits=n_splits)

    rf = RandomForestClassifier(
        random_state=RANDOM_STATE,
        n_jobs=-1,
        **rf_params
    )

    scoring = {"accuracy": "accuracy", "roc_auc": "roc_auc", "f1": "f1"}

    cv = cross_validate(
        rf, X_clean, Y_clean,
        cv=tscv,
        scoring=scoring,
        return_train_score=False
    )

    summary = {
        "model": model_name,
        "cv_accuracy_mean": cv["test_accuracy"].mean(),
        "cv_accuracy_std": cv["test_accuracy"].std(),
        "cv_roc_auc_mean": cv["test_roc_auc"].mean(),
        "cv_roc_auc_std": cv["test_roc_auc"].std(),
        "cv_f1_mean": cv["test_f1"].mean(),
        "cv_f1_std": cv["test_f1"].std(),
    }

    print(f"\n=== {model_name}: TimeSeries CV ({n_splits} folds) ===")
    print("CV Accuracy:  {:.4f} ± {:.4f}".format(summary["cv_accuracy_mean"], summary["cv_accuracy_std"]))
    print("CV ROC-AUC:   {:.4f} ± {:.4f}".format(summary["cv_roc_auc_mean"], summary["cv_roc_auc_std"]))
    print("CV F1-score:  {:.4f} ± {:.4f}".format(summary["cv_f1_mean"], summary["cv_f1_std"]))

    return summary


def plot_feature_importance(rf, feature_names, title):
    """
    Plot impurity-based feature importance (RF).
    Only meaningful when >=2 features.
    """
    importances = pd.Series(rf.feature_importances_, index=feature_names).sort_values()
    plt.figure(figsize=(6, 3))
    importances.plot(kind="barh")
    plt.title(title)
    plt.xlabel("Importance")
    plt.tight_layout()
    plt.show()
    return importances
```

### 4.4: define feature sets

```python
# RF-A: baseline
features_A = ["Price_Change_5d"]

# RF-B: + volatility regime
features_B = ["Price_Change_5d", "Volatility"]

# RF-C: + momentum oscillator
features_C = ["Price_Change_5d", "Volatility", "RSI"]

## RF-D control for price change and momentum oscillator
features_D = ["Price_Change_5d", "RSI"]
#consider adding MACD_Histogram + RSI + price_change_5d; BB_Width + RSI; RSI + BB_position

# RF-E control
features_E = ["Price_Change_5d", "RSI", "MACD_Histogram"]

features_F = ["Price_Change_5d", "RSI", "BB_Position"]

feature_sets = {
    "RF-A (Price_Change_5d)": features_A,
    "RF-B (Price_Change_5d+ Volatility)": features_B,
    "RF-C (Price_Change_5d+ Volatility + RSI)": features_C,
    "RF-D (Price_Change_5d + RSI)": features_D,
    "RF-E (Price_Change_5d+ RSI + MACD_Histogram)": features_E,
    "RF-F (Price_Change_5d + RSI + BB_Position)": features_F
}

# Use the tuned-style regularized RF settings you found helpful (shallow trees, larger leaf)
rf_params_regularized = {
    "n_estimators": 200,
    "max_depth": 3,
    "min_samples_leaf": 50,
    "max_features": 1.0
}
```

### 4.5: Train + Eval each feature set(hold out) + CV

```python
all_results = []
all_cv = []
all_importances = {}
RANDOM_STATE = 42

for name, feats in feature_sets.items():
    X = df[feats].copy()
    Y_named = Y.copy()
    Y_named.name = "Future_Up_5d"

    rf_model, res, X_clean, Y_clean = fit_eval_rf(
        X, Y_named, model_name=name, rf_params=rf_params_regularized
    )
    all_results.append(res)

    cv_summary = timeseries_cv_rf(
        X, Y_named, model_name=name, rf_params=rf_params_regularized, n_splits=5
    )
    all_cv.append(cv_summary)

    # feature importance only meaningful when >=2 features
    if len(feats) >= 2:
        imp = plot_feature_importance(rf_model, feats, title=f"Feature Importance - {name}")
        all_importances[name] = imp
```

### 4.6 Comparison table (hold-out + CV)

```python
# 1) convert collected outputs into DataFrames
results_df = pd.DataFrame(all_results)
cv_df = pd.DataFrame(all_cv)

print("results_df rows:", len(results_df))
print("cv_df rows:", len(cv_df))

# 2) merge hold-out + CV into one summary table
summary_df = results_df.merge(cv_df, on="model", how="outer", indicator=True)

# 3) quick diagnostic: check for missing models on either side
display(summary_df[["model", "_merge"]])

# 4) keep only fully matched rows for final comparison (both hold-out and CV exist)
summary_df = summary_df[summary_df["_merge"] == "both"].drop(columns=["_merge"])

# 5) create compact strings and sort by primary metric: CV ROC-AUC mean
summary_df["cv_roc_auc_mean_std"] = summary_df.apply(
    lambda r: f'{r["cv_roc_auc_mean"]:.4f} ± {r["cv_roc_auc_std"]:.4f}', axis=1
)

summary_df = summary_df.sort_values("cv_roc_auc_mean", ascending=False).reset_index(drop=True)

# 6) final clean view
summary_df_view = summary_df[[
    "model",
    "n_features",
    "features",
    "test_roc_auc",
    "cv_roc_auc_mean_std",
    "cv_roc_auc_mean",
    "cv_roc_auc_std",
    "cv_accuracy_mean",
    "cv_accuracy_std",
    "cv_f1_mean",
    "cv_f1_std"
]]

summary_df_view
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 4))
plt.bar(
    summary_df["model"],
    summary_df["cv_roc_auc_mean"],
    yerr=summary_df["cv_roc_auc_std"],
    capsize=5
)
plt.axhline(0.5, linestyle="--", linewidth=1)
plt.ylabel("CV ROC-AUC (mean ± std)")
plt.title("Random Forest Feature Augmentation Comparison")
plt.xticks(rotation=20, ha="right")
plt.tight_layout()
plt.show()
```

### 4.7 Interpretation
The feature augmentation experiments show that Random Forest performance improves only when additional indicators contribute complementary information beyond raw momentum. The baseline model using only Price_Change_5d (RF-A) performs close to random under time-series cross-validation (CV ROC-AUC ≈ 0.503), indicating that a single momentum feature contains only a weak predictive signal.

Adding Volatility (RF-B) does not improve results and increases variance across folds, suggesting that this volatility measure provides little stable incremental information for predicting 5-day direction in this dataset.

In contrast, adding RSI improves ranking performance (RF-D: CV ROC-AUC ≈ 0.514), indicating that oscillator-based momentum contributes additional, non-linear structure that the ensemble model can exploit.

The strongest performance is achieved when adding a third indicator that captures different dynamics. Both MACD_Histogram (RF-E) and BB_Position (RF-F) further improve cross-validated ROC-AUC to approximately 0.517, with RF-E also achieving the best hold-out ROC-AUC (≈ 0.531). This suggests that combining momentum (Price_Change_5d), oscillator signals (RSI), and trend-acceleration / regime information (MACD or band position) provides the most informative feature set among those tested.

Overall, while predictive power remains modest, the consistent uplift in CV ROC-AUC from RF-A/B to RF-D and further to RF-E/F shows that Random Forest benefits from carefully selected, complementary technical indicators rather than from adding features indiscriminately.

## Section 5: Conclusion
Across feature augmentations, Random Forest performance remains only moderately above random guessing, reflecting the inherent difficulty and noise of short-horizon stock direction prediction. However, the experiments clearly show that feature choice matters: volatility alone does not help, while momentum oscillators and trend/momentum-acceleration indicators provide consistent incremental value.

The best-performing feature set in this study is Price_Change_5d + RSI + MACD_Histogram (RF-E), which achieves the highest cross-validated ROC-AUC and the strongest hold-out ROC-AUC among tested configurations. This model is therefore selected as the final Random Forest specification for subsequent comparison against other model classes (e.g., deep learning).
